#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
df_users = pd.read_csv(r"users_table.csv")
df_users
```

#### 3. Rename Id column to userId

```python
df_users.rename(columns={"Id":"userId"}, inplace=True)
df_users
```

#### 4. Import posts table:

```python
df_posts = pd.read_csv(r"posts_table.csv")
df_posts
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
df_posts.rename(columns={"Id":"postId", "OwnerUserId":"userId"}, inplace=True)
df_posts
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
df_users = df_users.filter(["userId", "Reputation", "Views", "UpVotes", "DownVotes"], axis=1)
df_posts = df_posts.filter(["postId", "Score", "userId", "ViewCount", "CommentCount"], axis=1)
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
df_users_posts = df_users.merge(df_posts)
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
null_cols = df_users_posts.isnull().sum()
null_cols[null_cols > 0]
# Collumn ViewCount has 23572 missing/null values
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
# I will drop them because they make up a huge amount (> 10.000; more than half of total rows)
drop_cols = list(null_cols[null_cols > 10000].index)
df_users_posts = df_users_posts.drop(drop_cols, axis=1)
```

```python
df_users_posts
```

#### 10. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
df_users_posts.dtypes
# I would only change the column "Score" to float as a score could be a decimal number, 
# all of the other columns seem to be integers and so have correct data types
df_users_posts['Score'] = df_users_posts['Score'].astype('float')
df_users_posts['Score'].dtype
```
